 First of all we need to clean the data to make it ready to work . 


```python
# Data is need to be divided to quartz and jade roller parts. 
data = [] # define a list 
string = df['Campaign Name'].values #  save Campaign names in the string object
for st in string:
    y = "".join(itertools.takewhile(lambda x: x!=" ", st)) # extract the terms jad and quartz from the asscociated columns
    if y == 'Manual':
        y = 'Jade'
    data.append(y) # append each of the keywords to the list data
df['Product_type'] = data # add a new column product_type to the dataframe df
```

```python
# some of the terms are not in capital so we replace them with capital ones. 
df['Product_type'] = df['Product_type'].map(lambda x: x.replace('jade','Jade'),na_action='ignore')
df['Product_type'] = df['Product_type'].map(lambda x: x.replace('quartz','Quartz'),na_action='ignore')
```

### function to  find a new data frame based on the 3 important columns:

1. Customer Search Term
2. Spend
3. 7 Day Total Sales

we calculate sum of the sales and advertisement spends for each search terms which may be repeated in the data frame. 

```python

def Terms_ACoS(df):
    
    """
    INPUT:
    df - Pandas Data Frame 
    
    OUTPUT:
    df_new,df_sum_zero - The tuple for the new dataframe with two above mentioned columns and the the same dataframe with 
    zero sale. 
    
    """
    
    df_new = pd.DataFrame() # new data frame is defined
    df_new['sum_of_the_spends'] = df.groupby('Customer Search Term')['Spend'].sum() # sum of the all spends for each term is calculated 
    df_new['sum_of_the_sales'] = df.groupby('Customer Search Term')['7 Day Total Sales'].sum() # sum of the total sales for eac term is calculated
    
    df_sum_zero = df_new[df_new.sum_of_the_sales == 0] # sum of the sales which gives zero output are defined in the new data frame
    #df_new = df_new[df_new.sum_of_the_sales != 0]  # sum of the sales with non zero values are defined
    
    df_new['ACOS_mean'] = df_new['sum_of_the_spends']/df_new['sum_of_the_sales'] # new column is defined to calculate the ACOS(Total Advertising Cost of Sales)
    df_new = df_new.sort_values(ascending = True ,by = 'ACOS_mean') # we sort the values based on the column ACOS_mean in ascending order
    
    
    return df_new,df_sum_zero
```
